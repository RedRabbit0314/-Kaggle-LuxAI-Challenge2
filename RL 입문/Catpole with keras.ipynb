```python
import gym
import random
import os
import numpy as np
from collections      import deque
from keras.models     import Sequential
from keras.layers     import Dense
from keras.optimizers import Adam
from collections import deque
```

1. í•™ìŠµë¥ (learning rate):
- í•™ìŠµ ì‹œ ê°€ì¤‘ì¹˜(weight)ë¥¼ ì–¼ë§ˆë‚˜ ì¡°ì ˆí•  ê²ƒì¸ì§€ ê²°ì •í•˜ëŠ” ê°’
- ê°’ì´ í¬ë©´ í•™ìŠµì´ ë¹ ë¥´ê²Œ ì´ë£¨ì–´ì§€ì§€ë§Œ, overshooting ë¬¸ì œê°€ ë°œìƒí•  ìˆ˜ ìˆìŒ
- ê°’ì´ ì‘ìœ¼ë©´ í•™ìŠµì´ ëŠë¦¬ê²Œ ì´ë£¨ì–´ì§€ì§€ë§Œ, ì•ˆì •ì ìœ¼ë¡œ ìˆ˜ë ´í•˜ëŠ” ê²½í–¥ì´ ìˆìŒ

2. í• ì¸ìœ¨(gamma):
- ë¯¸ë˜ì˜ ë³´ìƒì„ í˜„ì¬ ë³´ìƒë³´ë‹¤ ì–¼ë§ˆë‚˜ ì¤‘ìš”í•˜ê²Œ ìƒê°í•  ê²ƒì¸ì§€ ê²°ì •í•˜ëŠ” ê°’
- ê°’ì´ 1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ë¯¸ë˜ì˜ ë³´ìƒì„ í˜„ì¬ ë³´ìƒê³¼ ë™ì¼í•˜ê²Œ ê°„ì£¼í•¨
- ê°’ì´ 0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ë¯¸ë˜ì˜ ë³´ìƒì„ í˜„ì¬ ë³´ìƒë³´ë‹¤ ëœ ì¤‘ìš”í•˜ê²Œ ê°„ì£¼í•¨

3. íƒí—˜ ë¹„ìœ¨(exploration rate):
- ì²˜ìŒì—ëŠ” íƒí—˜ì´ ë§ì´ ì´ë£¨ì–´ì§€ë‹¤ê°€ ì ì°¨ì ìœ¼ë¡œ íƒí—˜ì´ ì¤„ì–´ë“œëŠ” ê²ƒì„ ëª©í‘œë¡œ í•˜ëŠ” ê°’
- íƒí—˜ ë¹„ìœ¨ì´ ë†’ì„ìˆ˜ë¡ ë¬´ì‘ìœ„ë¡œ ë‹¤ì–‘í•œ í–‰ë™ì„ ì‹œë„í•˜ë¯€ë¡œ exploration-exploitation trade-offë¥¼ í•´ê²°í•˜ëŠ” ë° ë„ì›€ì´ ë¨

4. ìµœì†Œ íƒí—˜ ë¹„ìœ¨(exploration_min):
- íƒí—˜ ë¹„ìœ¨ì´ ì¤„ì–´ë“¤ì–´ì„œ ìµœì†Œ íƒí—˜ ë¹„ìœ¨ì— ë„ë‹¬í•˜ë©´ ê·¸ ì´í›„ë¶€í„°ëŠ” ë¬´ì¡°ê±´ ê°€ì¥ í° ë³´ìƒì„ ì£¼ëŠ” í–‰ë™ì„ ì„ íƒí•˜ë„ë¡ í•¨

5. íƒí—˜ ë¹„ìœ¨ ê°ì†Œìœ¨(exploration_decay):
- íƒí—˜ ë¹„ìœ¨ì„ ì–¼ë§ˆë‚˜ ë¹ ë¥´ê²Œ ê°ì†Œì‹œí‚¬ ê²ƒì¸ì§€ ê²°ì •í•˜ëŠ” ê°’
- ê°’ì´ ì‘ì„ìˆ˜ë¡ ëŠë¦¬ê²Œ ê°ì†Œí•˜ë¯€ë¡œ ë” ë§ì€ íƒí—˜ì„ ì‹œë„í•  ìˆ˜ ìˆìŒ
6. ì‹ ê²½ë§ ëª¨ë¸ ê°ì²´(brain):
- DQN ì•Œê³ ë¦¬ì¦˜ì—ì„œ Q í•¨ìˆ˜ë¥¼ ê·¼ì‚¬í•˜ê¸° ìœ„í•œ ì‹ ê²½ë§ ëª¨ë¸ ê°ì²´
- ì…ë ¥ìœ¼ë¡œ í˜„ì¬ ìƒíƒœ(state)ë¥¼ ë°›ì•„ì„œ ê° í–‰ë™(action)ì— ëŒ€í•œ Q ê°’ ì˜ˆì¸¡ê°’ì„ ì¶œë ¥í•¨

```python
class Agent():
    def __init__(self, state_size, action_size):
        self.weight_backup      = "cartpole_weight.h5" 
        self.state_size         = state_size 
        self.action_size        = action_size 
        self.memory             = deque(maxlen=2000) 
        self.learning_rate      = 0.001
        self.gamma              = 0.95
        self.exploration_rate   = 1.0
        self.exploration_min    = 0.01
        self.exploration_decay  = 0.995
        self.brain              = self._build_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(lr=self.learning_rate))

        if os.path.isfile(self.weight_backup): # ê°€ì¤‘ì¹˜ ë°±ì—… íŒŒì¼ì´ ì¡´ì¬í•˜ë©´, í•´ë‹¹ íŒŒì¼ì—ì„œ ë¯¸ë¦¬ í•™ìŠµëœ ê°€ì¤‘ì¹˜ë¥¼ ë¡œë“œ
            model.load_weights(self.weight_backup)
            self.exploration_rate = self.exploration_min
        return model

    def save_model(self):
            self.brain.save(self.weight_backup)

    def act(self, state):
        if np.random.rand() <= self.exploration_rate: # 0ê³¼ 1 ì‚¬ì´ì˜ ê· ì¼ ë¶„í¬ì—ì„œ ë‚œìˆ˜ë¥¼ ìƒì„± // íƒí—˜ì„ í†µí•´ ë¬´ì‘ìœ„ë¡œ í–‰ë™ì„ ì„ íƒ
            return random.randrange(self.action_size)
        
        act_values = self.brain.predict(state) # ì´ì „ì— í•™ìŠµí•œ ê²°ê³¼ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ê°€ì¥ ì¢‹ì€ í–‰ë™ì„ ì„ íƒ
        return np.argmax(act_values[0])

    def remember(self, state, action, reward, next_state, done): # next_state = í–‰ë™ìœ¼ë¡œ ì¸í•´ ì´ì–´ì§„ ë‹¤ìŒ ìƒíƒœ
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, sample_batch_size):
        if len(self.memory) < sample_batch_size:
            return
        sample_batch = random.sample(self.memory, sample_batch_size)
        for state, action, reward, next_state, done in sample_batch:
            target = reward
            if not done:
              target = reward + self.gamma * np.amax(self.brain.predict(next_state)[0])
            target_f = self.brain.predict(state)
            target_f[0][action] = target
            self.brain.fit(state, target_f, epochs=1, verbose=0)
        if self.exploration_rate > self.exploration_min: # íƒí—˜ ë¹„ìœ¨ì´ ìµœì†Œ íƒí—˜ ë¹„ìœ¨ë³´ë‹¤ ë†’ì€ ê²½ìš° decayë§Œí¼ ê°ì†Œ
            self.exploration_rate *= self.exploration_decay
```

```python
class CartPole:
    def __init__(self):
        self.sample_batch_size = 32 # í•™ìŠµí•  ë•Œ ì‚¬ìš©í•  ë°ì´í„° ìƒ˜í”Œì˜ í¬ê¸°
        self.episodes          = 10000
        self.env               = gym.make('CartPole-v1')
        
        self.state_size        = self.env.observation_space.shape[0] # self.env.observation_space.shapeëŠ” íŠœí”Œ, íŠœí”Œì€ ê° ì°¨ì›ì˜ í¬ê¸°ë¥¼ ë‚˜íƒ€ëƒ„
        # CartPole-v1: 4ê°€ì§€ ê°’ ê´€ì¸¡
        # ë§‰ëŒ€ì˜ ì†ë„, ê°ë„, ì¹´íŠ¸ì˜ ìœ„ì¹˜, ì¹´íŠ¸ì˜ ì†ë„
        # self.env.observation_space.shape[0] = (4,)        
        self.action_size       = self.env.action_space.n # nì„ ë¶™ì´ëŠ” ì´ìœ   = í•´ë‹¹ í™˜ê²½ì—ì„œ ê°€ëŠ¥í•œ actionì˜ ê°œìˆ˜ë¥¼ ì €ì¥
        self.agent             = Agent(self.state_size, self.action_size)


    def run(self):
        try:
            for index_episode in range(self.episodes):
                state, info = self.env.reset() 
                state = np.reshape(state, [1, self.state_size])

                done = False
                index = 0
                while not done:
#                   self.env.render()
                    action = self.agent.act(state)

                    next_state, reward, done, trunc, _ = self.env.step(action)
                    next_state = np.reshape(next_state, [1, self.state_size])
                    self.agent.remember(state, action, reward, next_state, done)
                    state = next_state
                    index += 1
                print("Episode {}# Score: {}".format(index_episode, index + 1)) # í˜„ì¬ ì—í”¼ì†Œë“œì˜ ë²ˆí˜¸ì™€ ì—ì´ì „íŠ¸ê°€ ì·¨í•œ ì•¡ì…˜ ìˆ˜ë¥¼ ì¶œë ¥
                self.agent.replay(self.sample_batch_size)
        finally:
            self.agent.save_model()
if __name__ == "__main__":
    cartpole = CartPole()
    cartpole.run()
```
