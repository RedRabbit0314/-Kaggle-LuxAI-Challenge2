```python
import gym
import random
import os
import numpy as np
from collections      import deque
from keras.models     import Sequential
from keras.layers     import Dense
from keras.optimizers import Adam
from collections import deque
```

1. 학습률(learning rate):
- 학습 시 가중치(weight)를 얼마나 조절할 것인지 결정하는 값
- 값이 크면 학습이 빠르게 이루어지지만, overshooting 문제가 발생할 수 있음
- 값이 작으면 학습이 느리게 이루어지지만, 안정적으로 수렴하는 경향이 있음

2. 할인율(gamma):
- 미래의 보상을 현재 보상보다 얼마나 중요하게 생각할 것인지 결정하는 값
- 값이 1에 가까울수록 미래의 보상을 현재 보상과 동일하게 간주함
- 값이 0에 가까울수록 미래의 보상을 현재 보상보다 덜 중요하게 간주함

3. 탐험 비율(exploration rate):
- 처음에는 탐험이 많이 이루어지다가 점차적으로 탐험이 줄어드는 것을 목표로 하는 값
- 탐험 비율이 높을수록 무작위로 다양한 행동을 시도하므로 exploration-exploitation trade-off를 해결하는 데 도움이 됨

4. 최소 탐험 비율(exploration_min):
- 탐험 비율이 줄어들어서 최소 탐험 비율에 도달하면 그 이후부터는 무조건 가장 큰 보상을 주는 행동을 선택하도록 함

5. 탐험 비율 감소율(exploration_decay):
- 탐험 비율을 얼마나 빠르게 감소시킬 것인지 결정하는 값
- 값이 작을수록 느리게 감소하므로 더 많은 탐험을 시도할 수 있음
6. 신경망 모델 객체(brain):
- DQN 알고리즘에서 Q 함수를 근사하기 위한 신경망 모델 객체
- 입력으로 현재 상태(state)를 받아서 각 행동(action)에 대한 Q 값 예측값을 출력함

```python
class Agent():
    def __init__(self, state_size, action_size):
        self.weight_backup      = "cartpole_weight.h5" 
        self.state_size         = state_size 
        self.action_size        = action_size 
        self.memory             = deque(maxlen=2000) 
        self.learning_rate      = 0.001
        self.gamma              = 0.95
        self.exploration_rate   = 1.0
        self.exploration_min    = 0.01
        self.exploration_decay  = 0.995
        self.brain              = self._build_model()

    def _build_model(self):
        # Neural Net for Deep-Q learning Model
        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(lr=self.learning_rate))

        if os.path.isfile(self.weight_backup): # 가중치 백업 파일이 존재하면, 해당 파일에서 미리 학습된 가중치를 로드
            model.load_weights(self.weight_backup)
            self.exploration_rate = self.exploration_min
        return model

    def save_model(self):
            self.brain.save(self.weight_backup)

    def act(self, state):
        if np.random.rand() <= self.exploration_rate: # 0과 1 사이의 균일 분포에서 난수를 생성 // 탐험을 통해 무작위로 행동을 선택
            return random.randrange(self.action_size)
        
        act_values = self.brain.predict(state) # 이전에 학습한 결과를 기반으로 가장 좋은 행동을 선택
        return np.argmax(act_values[0])

    def remember(self, state, action, reward, next_state, done): # next_state = 행동으로 인해 이어진 다음 상태
        self.memory.append((state, action, reward, next_state, done))

    def replay(self, sample_batch_size):
        if len(self.memory) < sample_batch_size:
            return
        sample_batch = random.sample(self.memory, sample_batch_size)
        for state, action, reward, next_state, done in sample_batch:
            target = reward
            if not done:
              target = reward + self.gamma * np.amax(self.brain.predict(next_state)[0])
            target_f = self.brain.predict(state)
            target_f[0][action] = target
            self.brain.fit(state, target_f, epochs=1, verbose=0)
        if self.exploration_rate > self.exploration_min: # 탐험 비율이 최소 탐험 비율보다 높은 경우 decay만큼 감소
            self.exploration_rate *= self.exploration_decay
```

```python
class CartPole:
    def __init__(self):
        self.sample_batch_size = 32 # 학습할 때 사용할 데이터 샘플의 크기
        self.episodes          = 10000
        self.env               = gym.make('CartPole-v1')
        
        self.state_size        = self.env.observation_space.shape[0] # self.env.observation_space.shape는 튜플, 튜플은 각 차원의 크기를 나타냄
        # CartPole-v1: 4가지 값 관측
        # 막대의 속도, 각도, 카트의 위치, 카트의 속도
        # self.env.observation_space.shape[0] = (4,)        
        self.action_size       = self.env.action_space.n # n을 붙이는 이유  = 해당 환경에서 가능한 action의 개수를 저장
        self.agent             = Agent(self.state_size, self.action_size)


    def run(self):
        try:
            for index_episode in range(self.episodes):
                state, info = self.env.reset() 
                state = np.reshape(state, [1, self.state_size])

                done = False
                index = 0
                while not done:
#                   self.env.render()
                    action = self.agent.act(state)

                    next_state, reward, done, trunc, _ = self.env.step(action)
                    next_state = np.reshape(next_state, [1, self.state_size])
                    self.agent.remember(state, action, reward, next_state, done)
                    state = next_state
                    index += 1
                print("Episode {}# Score: {}".format(index_episode, index + 1)) # 현재 에피소드의 번호와 에이전트가 취한 액션 수를 출력
                self.agent.replay(self.sample_batch_size)
        finally:
            self.agent.save_model()
if __name__ == "__main__":
    cartpole = CartPole()
    cartpole.run()
```
